```python
import numpy as np
import pandas as pd
import os

import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import matplotlib.pyplot as plt
import keras
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import cv2
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from keras.utils import to_categorical
from keras.models import Sequential
import tensorflow as tf
from keras.layers import Dense,Conv2D,Flatten,MaxPooling2D,Dropout
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/С3_04_NN_Intro/L04_Convolutional/data/train_images.csv')
```

```python
l1=[]
l2=[]
for img,ClassId,EncodedPixels in tqdm(data.values):
    try:
      # image=cv2.imread("train_images/{}".format(img), cv2.IMREAD_COLOR)
      # image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/С3_04_NN_Intro/L04_Convolutional/data/train_images/{}".format(img),cv2.IMREAD_COLOR)
      image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/С3_04_NN_Intro/L04_Convolutional/data/train_images/{}".format(img),cv2.IMREAD_GRAYSCALE)
      image = cv2.resize(image, (120, 120))
      l1.append(image)
      l2.append(ClassId)
    except:
      continue
```

```python
encoder = LabelEncoder()

X= np.array(l1)
X = X/255

y = encoder.fit_transform(l2)
y = to_categorical(y)
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,
                                                  y,
                                                  test_size=0.2,
                                                  stratify=y,
                                                  shuffle=True)
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(120, 120, 1), activation="relu"))
model.add(MaxPooling2D(pool_size=(3, 3)))
model.add(Conv2D(64, (3, 3), activation="relu"))
model.add(MaxPooling2D (pool_size=(3, 3)))
model.add(Conv2D(64, (3, 3), activation="relu"))
model.add(MaxPooling2D(pool_size=(4, 4)))
model.add(Flatten())
model.add(Dense(128, activation="relu"))
model.add(Dropout(0.3))
model.add(Dense(128, activation="relu"))
model.add(Dropout(0.3))
model.add(Dense(256, activation="relu"))
model.add(Dense(4, activation="softmax"))
```

```python
model.compile(loss=keras.losses.categorical_crossentropy,
             optimizer=keras.optimizers.Adam(),
             metrics=["accuracy"])
```

```python
history = model.fit(X_train,
                    y_train,
                    epochs=15,
                    validation_data=(X_test, y_test),
                    batch_size=128,
                    verbose=1)
```

```python
def rle_encode(mask):
    rle_encoded_mask = []
    prev_value = mask[0][0]  # Access first element
    count = 1

    for value in mask[1:]:
        if value.any() != prev_value:  # Check if any element in value differs from prev_value
            rle_encoded_mask.append(int(prev_value))
            prev_value = value[0]  # Access first element for next comparison
            count = 1
        else:
            count += 1

        rle_encoded_mask.append(int(prev_value))
    return ' '.join(map(str, rle_encoded_mask))
```

```python
test_images_dir = "/content/drive/MyDrive/Colab Notebooks/С3_04_NN_Intro/L04_Convolutional/data/test_images/"
test_filenames = os.listdir(test_images_dir)
```

```python
y_pred = []
encoded_pixels = []
```

```python
for filename in tqdm(test_filenames):
    # Загрузка и преобразование изображения
    image = cv2.imread(os.path.join(test_images_dir, filename), cv2.IMREAD_GRAYSCALE)
    image = cv2.resize(image, (120, 120))
    _, mask = cv2.threshold(image, thresh=127, maxval=255, type=cv2.THRESH_BINARY)

    # RLE-кодирование маски
    rle_encoded_mask = rle_encode(mask)
    encoded_pixels.append(rle_encoded_mask)

    # Предикция класса
    prediction = model.predict(np.expand_dims(image, axis=0))
    predicted_class = np.argmax(prediction, axis=1)[0] + 1
    y_pred.append(predicted_class)

# Создание dataframe
submission_df = pd.DataFrame({
    "ImageId": test_filenames,
    "EncodedPixels": encoded_pixels,
    "ClassId": y_pred
})

# Сохранение dataframe без заголовка в файл submission.csv
submission_df.to_csv("/content/drive/MyDrive/Colab Notebooks/С3_04_NN_Intro/Graduation/submission.csv", index=False, header=False)
```
